```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/Internetusage_Beginnertask03.csv")

df['SessionStart'] = pd.to_datetime(df['SessionStart'])
df['SessionEnd'] = pd.to_datetime(df['SessionEnd'])

df['hour'] = df['SessionStart'].dt.hour
df['duration_min'] = (
    df['SessionEnd'] - df['SessionStart']
).dt.total_seconds() / 60

df['transfer_rate'] = df['DataTransferred'] / df['duration_min']
df['wifi_load'] = df['transfer_rate'] * df['duration_min']
```

```python
# Peak usage hours
df.groupby('hour')['DataTransferred'].sum().plot(kind='bar')
plt.title("Total Data Usage by Hour")
plt.show()
```

```python
# Heavy users
df.groupby('UserID')['DataTransferred'].sum().sort_values(ascending=False).head(10)
```

```python
# Idle timeout sessions
df[df['Status'] == 'Idle Timeout'][['duration_min', 'DataTransferred']].describe()
```
